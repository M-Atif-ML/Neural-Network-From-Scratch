## Lecture 1

```python
inputs = [1,2,3]
weights = [0.2,0.8,-0.5]
bias=2

output = (inputs[0]*weights[0]+inputs[1]*weights[1]+inputs[2]*weights[2] + bias)
output
```

```python
inputs = [1,2,3,2.5]
weights = [0.2,0.8,-0.5,1.0]
bias=2

output = (inputs[0]*weights[0]+inputs[1]*weights[1]+inputs[2]*weights[2]+inputs[3]*weights[3] + bias)
output
```

```python
inputs = [1,2,3,2.5]
weights = [[0.2,0.8,-0.5,1],       
          [0.5,-0.91,0.26,-0.5],
          [-0.26,-0.27,0.17,0.87]]
biases = [2,3,0.5]
weights1 = weights[0] # List of weights associated with 1st neuron: w11,w12,w13,w14
weights2 = weights[1] # List of weights associated with 2nd neuron: w12,w22,w23,w24 
weights3 = weights[2] # List of weights associated with 3rd neuron: w13,w23,w33,w34

outputs = [
    inputs[0]*weights1[0]+inputs[1]*weights1[1]+inputs[2]*weights1[2]+inputs[3]*weights1[3]+biases[0],
    inputs[0]*weights2[0]+inputs[1]*weights2[1]+inputs[2]*weights2[2]+inputs[3]*weights2[3]+biases[1],
    inputs[0]*weights3[0]+inputs[1]*weights3[1]+inputs[2]*weights3[2]+inputs[3]*weights3[3]+biases[2],
]
outputs
```

```python
inputs = [1.1,2.2,3.3]

weights = [[2,0,0],
          [2,2,0],
          [2,2,2]]

biases = [0,0,1]


layer_outputs = []

for weight,biase in zip(weights,biases):
    output = 0

    for w,i in zip(weight,inputs):
        output += w * i
    output+=biase
    layer_outputs.append(output)

layer_outputs
```

### Dot product

```python
inputs = np.array([1.0,2.0,3.0,2.5])
weights = np.array([[0.2,0.8,-0.5,1],
           [0.5,-0.91,0.26,-0.5],
           [-0.26,-0.27,0.17,0.87]])
biases = [2.0,3.0,0.5]

layer_outputs = np.dot(weights,inputs)+biases
print(f"weights.inputs:{layer_outputs}")


layer_outputs = np.dot(inputs,weights.T)+biases
print(f"inputs.transpose(weights):{layer_outputs}")
    
```

```python
inputs = np.array([[1.0,2.0,3.0,2.5],[2.0,5.0,-1.0,2.0],[-1.5,2.7,3.3,-0.8]])
weights = np.array([[0.2,0.8,-0.5,1],
                   [0.5,-0.91,0.26,-0.5],
                   [-0.26,-0.27,0.17,0.87]])
biases = [2.0,3.0,0.5]
 
layer_outputs = np.dot(inputs,weights.T)
layer_outputs
```
